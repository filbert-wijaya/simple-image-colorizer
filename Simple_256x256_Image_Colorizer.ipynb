```python
import tensorflow as tf
import tensorflow_io as tfio
import matplotlib.pyplot as plt
```

```python
model = tf.keras.models.load_model('image_colorizer.keras', safe_mode=False)
```

```python
# Image 1
i = 1

# Image 2
# i = 2

# Image 3
# i = 3

# Image 4
# i = 4

# Image 5
# i = 5
```

```python
image = tf.keras.utils.load_img(f'image{i}.jpg')
input_image = tf.keras.utils.img_to_array(image)
input_image = tf.image.resize(input_image, (256,256))
input_image = tf.reshape(input_image, [1, 256, 256, 3])
result = model.predict(input_image)
```

```python
plt.imshow(tf.squeeze(input_image) / 255)
plt.show()
```

```python
plt.imshow(result)
plt.show()
```

```python
plt.subplots_adjust(wspace=0.4, hspace=0.4)

plt.subplot(1,3,1)
plt.title('Input')
plt.imshow(tf.squeeze(input_image) / 255)

plt.subplot(1,3,2)
plt.title('Output')
plt.imshow(result)

image = tf.keras.utils.load_img(f'image{i}_color.jpg')
color_image = tf.keras.utils.img_to_array(image)

plt.subplot(1,3,3)
plt.title('Color Image')
plt.imshow(color_image / 255)

plt.show()
```
